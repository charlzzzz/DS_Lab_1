```python
import pandas as pd
import ta
from itertools import islice

# âœ… Rename columns to lowercase for consistency
data.rename(columns={
    'Close': 'close',
    'High': 'high',
    'Low': 'low',
    'Open': 'open',
    'Adj Close': 'adj_close',
    'Volume': 'volume',
    'Ticker': 'ticker'
}, inplace=True)

features = []

def chunked(iterable, size):
    it = iter(iterable)
    return iter(lambda: list(islice(it, size)), [])

def add_rolling_and_lagged_features(df):
    lag_features = {
        "rsi": [1, 5],
        "macd": [1],
        "macd_hist": [1, 5],
        "price_ema_dist": [5],
        "adx": [1, 5],
        "cci": [1, 5],
        "stoch_k": [1, 5],
       # "williams_r": [1, 5],
    }

    rolling_features = {
        "rsi": [5],
        "macd_hist": [5, 10],
        "price_ema_dist": [5],
        "obv": [5],
        "atr": [5],
        "adx": [5],
        "cci": [5],
        "stoch_k": [5]
      #  "williams_r": [5]
    }

    for feature, lags in lag_features.items():
        if feature in df.columns:
            for lag in lags:
                df[f"{feature}_lag{lag}"] = df[feature].shift(lag)

    for feature, windows in rolling_features.items():
        if feature in df.columns:
            for win in windows:
                df[f"{feature}_roll_mean_{win}"] = df[feature].rolling(win).mean().shift(1)
                #df[f"{feature}_roll_std_{win}"] = df[feature].rolling(win).std().shift(1)
                #df[f"{feature}_roll_min_{win}"] = df[feature].rolling(win).min().shift(1)
                #df[f"{feature}_roll_max_{win}"] = df[feature].rolling(win).max().shift(1)

    return df

symbols = data['ticker'].unique()

for batch_num, batch in enumerate(chunked(symbols, 50), start=1):
    print(f"Processing batch {batch_num} with {len(batch)} symbols...")
    batch_data = data[data['ticker'].isin(batch)].copy()
    
    for symbol in batch:
        print(f" â†’ Processing symbol: {symbol}")
        df_symbol = batch_data[batch_data['ticker'] == symbol].copy()
        
        if df_symbol.shape[0] < 30:
            print(f" âš  Skipping {symbol}: insufficient data ({df_symbol.shape[0]} rows)")
            continue
        
        # === Existing indicators ===
        df_symbol['rsi'] = ta.momentum.RSIIndicator(close=df_symbol['close'], window=14).rsi()
        ema = ta.trend.EMAIndicator(close=df_symbol['close'], window=20)
        df_symbol['ema'] = ema.ema_indicator()
        df_symbol['price_ema_dist'] = ((df_symbol['close'] - df_symbol['ema']) / df_symbol['ema']) * 100
        macd = ta.trend.MACD(close=df_symbol['close'])
        df_symbol['macd'] = macd.macd()
        df_symbol['macd_hist'] = macd.macd_diff()
        bb = ta.volatility.BollingerBands(close=df_symbol['close'])
        df_symbol['bb_upper'] = bb.bollinger_hband()
        df_symbol['bb_lower'] = bb.bollinger_lband()
        df_symbol['obv'] = ta.volume.OnBalanceVolumeIndicator(close=df_symbol['close'], volume=df_symbol['volume']).on_balance_volume()
        df_symbol['atr'] = ta.volatility.AverageTrueRange(high=df_symbol['high'], low=df_symbol['low'], close=df_symbol['close']).average_true_range()
        df_symbol['return_5d'] = df_symbol['close'].pct_change(periods=5) * 100

        # === ðŸš€ New indicators ===
        df_symbol['adx'] = ta.trend.ADXIndicator(high=df_symbol['high'], low=df_symbol['low'], close=df_symbol['close'], window=14).adx()
        df_symbol['cci'] = ta.trend.CCIIndicator(high=df_symbol['high'], low=df_symbol['low'], close=df_symbol['close'], window=14).cci()
        stoch = ta.momentum.StochasticOscillator(high=df_symbol['high'], low=df_symbol['low'], close=df_symbol['close'])
        df_symbol['stoch_k'] = stoch.stoch()
        df_symbol['ultimate_osc'] = ta.momentum.UltimateOscillator(high=df_symbol['high'], low=df_symbol['low'], close=df_symbol['close']).ultimate_oscillator()
        df_symbol['cmf'] = ta.volume.ChaikinMoneyFlowIndicator(high=df_symbol['high'], low=df_symbol['low'], close=df_symbol['close'], volume=df_symbol['volume']).chaikin_money_flow()

        for ind in ['rsi', 'macd', 'macd_hist', 'adx', 'cci', 'stoch_k']:
            if ind in df_symbol.columns:
                df_symbol[f'{ind}_diff'] = df_symbol[ind].diff()

        df_symbol['close_diff']   = df_symbol['close'].diff()               # Î” close
        df_symbol['high_diff']    = df_symbol['high'].diff()
        df_symbol['low_diff']     = df_symbol['low'].diff()
        df_symbol['volume_diff']  = df_symbol['volume'].diff()

        df_symbol['close_pct']    = df_symbol['close'].pct_change() * 100   # %Î”
        df_symbol['high_pct']     = df_symbol['high'].pct_change()  * 100
        df_symbol['low_pct']      = df_symbol['low'].pct_change()   * 100
        df_symbol['volume_pct']   = df_symbol['volume'].pct_change() * 100

        # === ðŸ†• Longer-horizon & vol features (paste here) ===
        #df_symbol['return_10d']  = df_symbol['close'].pct_change(10) * 100
        df_symbol['return_20d']  = df_symbol['close'].pct_change(20) * 100
        #df_symbol['volatility_10d'] = df_symbol['close'].pct_change().rolling(10).std() * 100
        df_symbol['volatility_20d'] = df_symbol['close'].pct_change().rolling(20).std() * 100

        # Add lagged + rolling features
        df_symbol = add_rolling_and_lagged_features(df_symbol)
        
        # Target
        #df_symbol['target'] = (df_symbol['close'].shift(-1) > df_symbol['close']).astype(int)

        threshold = 0.5  # Example: 0.5% movement considered significant
        df_symbol['pct_change'] = (df_symbol['close'].shift(-1) - df_symbol['close']) / df_symbol['close'] * 100

        df_symbol['target'] = df_symbol['pct_change'].apply(
            lambda x: 1 if x > threshold else (0 if x < -threshold else -1)
        )
        
        df_symbol = df_symbol[df_symbol['target'] != -1] # drop pct change so that the model doesn't see future data
        df_symbol.drop(columns=['pct_change'], inplace=True)


        # Drop NaNs
        df_symbol.dropna(inplace=True)
        
        # Collect processed symbol
        features.append(df_symbol)

# Combine all results
final_df = pd.concat(features, ignore_index=True)
final_df["ticker_id"] = final_df["ticker"].astype("category").cat.codes

# Save to CSV
output_file = 'features_expanded.csv'
final_df.to_csv(output_file, index=False)
print(f"\nâœ… Saved final dataset to '{output_file}' with {final_df.shape[0]} rows and {final_df.shape[1]} columns.")

# Show preview and count of final dataframe
print(final_df.head())
print(final_df.count())
```

```python
# -------------------------------------------------------------------------
# 0. Imports (only keep ONE copy of these in your notebook)
# -------------------------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import optuna
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (
    roc_auc_score, accuracy_score, matthews_corrcoef, classification_report
)

# -------------------------------------------------------------------------
# 1. DATA CLEAN-UP  (âœ“ this is where the old NaN-handling used to be)
# -------------------------------------------------------------------------
# 1.1  Build feature matrix and target vector
X_full = features.copy().reset_index(drop=True)          # <-- features comes from earlier cell
y_full = final_df["target"].copy().reset_index(drop=True)  # <-- target column

# 1.2  Drop any row that has a NaN in *any* predictor
valid_mask   = X_full.notna().all(axis=1)                # Boolean mask
X            = X_full.loc[valid_mask].reset_index(drop=True)
y            = y_full.loc[valid_mask].reset_index(drop=True)

# 1.3  Remember which original rows survived (for write-back later)
valid_indices = X_full.index[valid_mask]                 # <-- CRITICAL!

assert len(X) == len(y)

# -------------------------------------------------------------------------
# 2. CHRONOLOGICAL TRAIN / TEST SPLIT  (80 % / 20 %)
# -------------------------------------------------------------------------
split_idx   = int(len(X) * 0.80)
X_trainval, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_trainval, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# -------------------------------------------------------------------------
# 3. OPTUNA + TIME-SERIES CV
# -------------------------------------------------------------------------
tscv = TimeSeriesSplit(n_splits=5)   # expanding window

def objective(trial):
    params = {
        "n_estimators"     : trial.suggest_int("n_estimators", 300, 1200),
        "max_depth"        : trial.suggest_int("max_depth", 3, 10),
        "learning_rate"    : trial.suggest_float("learning_rate", 0.01, 0.2),
        "subsample"        : trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree" : trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "gamma"            : trial.suggest_float("gamma", 0, 5),
        "lambda"           : trial.suggest_float("lambda", 1e-3, 10, log=True),
        "alpha"            : trial.suggest_float("alpha", 1e-3, 10, log=True),
        "min_child_weight" : trial.suggest_int("min_child_weight", 1, 10),
        "random_state"     : 42,
        "eval_metric"      : "auc",
        "tree_method"      : "hist",
    }

    aucs = []
    for tr_idx, val_idx in tscv.split(X_trainval):
        X_tr, X_val = X_trainval.iloc[tr_idx], X_trainval.iloc[val_idx]
        y_tr, y_val = y_trainval.iloc[tr_idx], y_trainval.iloc[val_idx]

        model = XGBClassifier(**params)
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_val, y_val)],
            verbose=False,
        )

        proba_val = model.predict_proba(X_val)[:, 1]
        aucs.append(roc_auc_score(y_val, proba_val))

    return np.mean(aucs)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50, show_progress_bar=False)

print("âœ… Best params:", study.best_params)
print(f"âœ… Best CV AUC: {study.best_value:.4f}")

# -------------------------------------------------------------------------
# 4. FINAL MODEL  (train on full trainval, test on hold-out)
# -------------------------------------------------------------------------
best_model = XGBClassifier(
    enable_categorical=True,
    booster="gbtree",
    random_state=42,
    eval_metric="auc",
    tree_method="hist",
    **study.best_params,
)
best_model.fit(
    X_trainval,
    y_trainval,
    eval_set=[(X_test, y_test)],
    verbose=False,
)

y_pred     = best_model.predict(X_test)
proba_test = best_model.predict_proba(X_test)[:, 1]

print(f"\nâœ… Hold-out Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"âœ… Hold-out AUC      : {roc_auc_score(y_test, proba_test):.4f}")
print(f"âœ… Hold-out MCC      : {matthews_corrcoef(y_test, y_pred):.4f}")
print("\nâœ… Classification report:")
print(classification_report(y_test, y_pred))

plot_importance(best_model, max_num_features=20)
plt.show()

# -------------------------------------------------------------------------
# 5. WRITE PREDICTION PROBABILITIES BACK TO *features*
# -------------------------------------------------------------------------
features["xgb_pred_prob"] = np.nan                     # create the column
features.loc[valid_indices, "xgb_pred_prob"] = best_model.predict_proba(X)[:, 1]

assert not features.loc[valid_indices, "xgb_pred_prob"].isna().any()
print("\nâœ… Added 'xgb_pred_prob' to original features.")

# -------------------------------------------------------------------------
# 6. SAVE
# -------------------------------------------------------------------------
features.to_csv("features_with_xgb_pred.csv", index=False)
print("âœ… Saved to 'features_with_xgb_pred.csv'")
```

```python
import gym
from gym import spaces
import numpy as np
import pandas as pd


class TradingEnv(gym.Env):
    """
    A simple stock-trading environment for PPO.
    """
    metadata = {"render.modes": ["human"]}

    def __init__(self, data: pd.DataFrame, initial_balance: float = 10_000):
        super().__init__()

        self.data = data.reset_index(drop=True)
        self.initial_balance = initial_balance
        self.np_random = None

        # ---------- internal state ----------
        self.current_step = 0
        self.balance = initial_balance
        self.position = 0              # 0â€Š=â€Šflat, 1â€Š=â€Šlong, -1â€Š=â€Šshort
        self.entry_price = 0
        self.last_action = None

        ### NEW (Item 2) â€“ additional tracking  #########
        self.steps_in_position = 0      # how long weâ€™ve held the position
        ##################################################

        # ---------- action / obs spaces ----------
        self.action_space = spaces.Discrete(3)     # 0 Hold  1 Buy  2 Sell

        sample_obs = self._get_obs()
        self.observation_space = spaces.Box(
            low=-np.inf,
            high=np.inf,
            shape=sample_obs.shape,
            dtype=np.float32,
        )

    # ------------------------------------------------------------------
    #  observation
    # ------------------------------------------------------------------
    def _get_obs(self) -> np.ndarray:
        row = self.data.iloc[self.current_step]

        core_feats = row[
            ["xgb_pred_prob", "adx", "price_ema_dist_lag5", "volatility_20d"]
        ].astype(np.float32).to_numpy()

        ### NEW (Item 3) â€“ simple 5-step momentum window ###############
        # five recent percentage returns; zero-padded for early steps
        mom = np.zeros(5, dtype=np.float32)
        for i in range(1, 6):
            idx = self.current_step - i
            if idx >= 0:
                close_now  = row["close"]
                close_prev = self.data.iloc[idx]["close"]
                mom[i - 1] = (close_now / close_prev) - 1.0
        ################################################################

        # position info
        position = np.array([self.position], dtype=np.float32)
        normalized_entry = (
            np.array([self.entry_price / row["close"]], dtype=np.float32)
            if self.entry_price > 0
            else np.array([0.0], dtype=np.float32)
        )

        ### NEW (Item 2) â€“ steps-in-pos feature ########################
        pos_state = np.array([self.steps_in_position], dtype=np.float32)
        ################################################################

        obs = np.concatenate(
            [core_feats, mom, position, normalized_entry, pos_state]
        )

        ### NEW â€“ NaN sanitiser (prevents nan-logit crash) ##############
        obs = np.nan_to_num(obs, nan=0.0, posinf=0.0, neginf=0.0)
        ################################################################
        return obs

    # ------------------------------------------------------------------
    #  reset
    # ------------------------------------------------------------------
    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        self.balance = self.initial_balance
        self.position = 0
        self.entry_price = 0
        ### NEW (Item 2)
        self.steps_in_position = 0
        ###
        return self._get_obs(), {}
    
    def render(self, mode="human"):
        print(
            f"Step {self.current_step:4} | "
            f"Bal: {self.balance:,.2f} | Pos: {self.position:+d} | "
            f"Steps-in-pos: {self.steps_in_position}"
        )

    def seed(self, seed=None):
        self.np_random, seed = gym.utils.seeding.np_random(seed)
        return [seed]

    def step(self, action):
        done = False
        reward = 0

        prev_action    = self.last_action
        profit_pct     = 0.0
        starting_balance = self.balance

        TX_COST = 0.003
        SCALE_PCT = 12.0
        HOLDING_REWARD = 0.001
        #SMALL_PROFIT_THRESHOLD = 0.005
        #SMALL_PROFIT_PENALTY = 0.2
        current_price = self.data.iloc[self.current_step]["close"]
        

        # ----------- redundant-action penalties -----------------
        # if action == 1 and self.position == 1:
        #     reward -= 0.05
        # if action == 2 and self.position == -1:
        #     reward -= 0.05
        # --------------------------------------------------------

         # ðŸ’¡ Add idle-Hold penalty 
        # if action == 0 and self.position == 0:
        #     reward -= 0.03   # mild penalty for being inactive

        # ----------------- action handling ----------------------
        if action == 1:                                   # BUY
            if self.position == 0:                        # open LONG
                self.position = 1
                self.entry_price = current_price
                self.entry_step  = self.current_step
                reward -= TX_COST
                
            elif self.position == -1:                     # close SHORT
                profit_pct = (self.entry_price - current_price) / self.entry_price
                reward    += SCALE_PCT * profit_pct - TX_COST

                # if abs(profit_pct) < SMALL_PROFIT_THRESHOLD:
                #     reward -= SMALL_PROFIT_PENALTY
                # elif abs(profit_pct) > 0.015:
                #     reward += 1.0 * abs(profit_pct)

                  # SIMPLIFIED: Balanced large trade handling
            if profit_pct > 0.02:
                reward += 0.5  # Fixed bonus for good trades
            elif profit_pct < -0.02:
                reward -= 1.0  # Fixed penalty for bad trades

            # REMOVED: Flip-flop penalties
            
            self.balance *= (1 + profit_pct)
            self.position = 0 # flat

        elif action == 2 and self.position == 1:          # close LONG
            profit_pct = (current_price - self.entry_price) / self.entry_price
            reward    += SCALE_PCT * profit_pct - TX_COST

            # SIMPLIFIED: Same as SHORT closing
            if profit_pct > 0.02:
                reward += 0.5
            elif profit_pct < -0.02:
                reward -= 1.0

            self.balance *= (1 + profit_pct)
            self.position = 0

            # if abs(profit_pct) < SMALL_PROFIT_THRESHOLD:
            #     reward -= SMALL_PROFIT_PENALTY
            # elif abs(profit_pct) > 0.015:
            #     reward += 1.0 * abs(profit_pct)
            # if profit_pct < 0.015:
            #     reward -= 2.0 * abs(profit_pct)

            # if self.current_step - self.entry_step <= 3:
            #     reward -= 0.05

            # self.balance *= (1 + profit_pct)
            # self.position = 0

        elif action == 2 and self.position == 0:          # open SHORT
            self.position = -1
            self.entry_price = current_price
            self.entry_step  = self.current_step
            reward -= TX_COST

         # NEW: Reward for holding positions (encourages more trading)
        if self.position != 0:
            reward += HOLDING_REWARD
        

        # unrealised PnL drip
        if self.position == 1:
            unreal = (current_price - self.entry_price) / self.entry_price
            reward += 0.003 * unreal
        elif self.position == -1:
            unreal = (self.entry_price - current_price) / self.entry_price
            reward += 0.003 * unreal

            # NEW: Time-based position closing pressure
        if self.position != 0:
            hold_duration = self.current_step - self.entry_step
            if hold_duration > 20:  # After 20 steps, start encouraging closure
                reward -= 0.001 * (hold_duration - 20)


        reward = float(np.clip(reward, -3, 3))  # Increased range from -3,3

        # advance
        self.current_step += 1
        if self.current_step >= len(self.data) - 1:
            done = True

        # ------------ realised PnL (percentage) -----------------
        realised_pct = (
            (self.balance / starting_balance) - 1.0
            if self.balance != starting_balance else 0.0
        )                                                 # <<< REPLACED block
        # --------------------------------------------------------

        info = {
            "balance":   self.balance,
            "trade_pnl": realised_pct,
            "position":  self.position,
            "entry_price": self.entry_price,
        }

        self.last_action = action
        return self._get_obs(), reward, done, False, info 
```

```python
from stable_baselines3 import PPO
import random
import tensorboard
from stable_baselines3.common.utils import get_linear_fn
from stable_baselines3.common.callbacks import BaseCallback
from stable_baselines3.common.vec_env import DummyVecEnv,VecMonitor, VecNormalize
from gym.wrappers import TimeLimit


# ---------- Ground-truth action labeler -----------------------------------
def generate_ground_truth_actions(df, threshold, lookahead):
    """
    0 = HOLD, 1 = BUY, 2 = SELL
    BUY  if future return  > +threshold
    SELL if future return  < â€“threshold
    else HOLD
    """
    future_ret = (df['close'].shift(-lookahead) - df['close']) / df['close']

    def rule(r):
        if r >  threshold:  return 1
        if r < -threshold:  return 2
        return 0

    return future_ret.apply(rule).fillna(0).astype(int).to_numpy()
# --------------------------------------------------------------------------

# ------------------------------------------------------------
# Better ground-truth generator  (keeps the old one untouched)
# ------------------------------------------------------------
def create_better_baseline_labels(
        df: pd.DataFrame,
        price_col: str = "close",
        volatility_threshold: float = 0.02,
        momentum_window: int = 5
    ) -> np.ndarray:
    """
    0 = Hold, 1 = Buy, 2 = Sell
    Combines momentum, low-volatility filter and an RSI-like gauge.
    """
    prices = df[price_col].values
    n = len(prices)
    labels = np.zeros(n, dtype=int)

    # ---- helpers ---------------------------------------------------------
    returns = np.diff(prices, prepend=prices[0]) / prices  # 1-step returns
    vol     = pd.Series(returns).rolling(20).std().fillna(0).values

    # momentum over a window
    mom = np.zeros(n)
    mom[momentum_window:] = (prices[momentum_window:] - prices[:-momentum_window]) / prices[:-momentum_window]

    # simple RSI approximation
    gains  = np.clip(returns, 0, None)
    losses = np.clip(-returns, 0, None)
    avg_g  = pd.Series(gains ).rolling(14).mean().fillna(0).values
    avg_l  = pd.Series(losses).rolling(14).mean().fillna(0).values
    rsi    = 100 - (100 / (1 + avg_g / (avg_l + 1e-8)))

    # ---- rule set --------------------------------------------------------
    for i in range(n):
        if (
            mom[i] > 0.03 and             # strong up momentum
            vol[i] < volatility_threshold and
            rsi[i] < 70
        ):
            labels[i] = 1                 # BUY
        elif (
            mom[i] < -0.03 and            # strong down momentum
            vol[i] < volatility_threshold and
            rsi[i] > 30
        ):
            labels[i] = 2                 # SELL
        else:
            labels[i] = 0                 # HOLD
    return labels



lr_schedule = get_linear_fn(2e-4, 5e-5, 750000)

# Create environment
#small_features = features.head(100000)
features = pd.read_csv("features_with_xgb_pred.csv")
#print("checking date order in final_df ")
final_df = final_df.sort_values('Date').reset_index(drop=True)
assert final_df['Date'].is_monotonic_increasing
#print ("Date order is correct in final_df")

env_df = (
    final_df.sort_values("Date")
            .reset_index(drop=True)
            [["Date", "close"]] 
            .join(features)
)

TEST_RATIO   = 0.20
split_idx    = int(len(env_df) * (1 - TEST_RATIO))

train_df_raw = env_df.iloc[:split_idx].dropna().reset_index(drop=True)  
test_df_raw  = env_df.iloc[split_idx:].dropna().reset_index(drop=True) 

# -------- Ground-truth labels (test set only) -----------------------------
#gt_actions_test = generate_ground_truth_actions(test_df_raw,threshold=0.05,lookahead=3)

gt_actions_test = create_better_baseline_labels(test_df_raw)

from collections import Counter


print("GT label distribution:", Counter(gt_actions_test))


non_numeric = ["Date"]
train_df = train_df_raw.drop(columns=non_numeric)
test_df  = test_df_raw.drop(columns=non_numeric)

assert train_df.select_dtypes("number").shape[1] == train_df.shape[1]
assert not train_df.isna().any().any()
assert not test_df.isna().any().any()

assert final_df['Date'].is_monotonic_increasing
def make_env(df):
    return TimeLimit(TradingEnv(df), max_episode_steps=240)

train_env_raw = DummyVecEnv([lambda: make_env(train_df)])
test_env_raw  = DummyVecEnv([lambda: make_env(test_df)])

# â¬‡ï¸ Normalize observations (not rewards)
train_env = VecNormalize(train_env_raw, norm_obs=True, norm_reward=False, clip_obs=10.0)
test_env  = VecNormalize(test_env_raw,  norm_obs=True, norm_reward=False, training=False)

# Add monitor wrappers for logging
train_env = VecMonitor(train_env)
test_env  = VecMonitor(test_env)


#env = VecMonitor(train_env)
SEED = 42
random.seed(SEED)


class TradeLogger(BaseCallback):
    def __init__(self, log_every=20, verbose=0):
        super().__init__(verbose)
        self.log_every = log_every
        self.episodes = []
        self.ep_pnl = 0

    def _on_step(self) -> bool:
        # Info dicts come back in a list (vec-env)
        info = self.locals["infos"][0]
        if "trade_pnl" in info:          # weâ€™ll add this key in env.step()
            self.ep_pnl += info["trade_pnl"]

        if self.locals["dones"][0]:      # episode finished
            self.episodes.append(self.ep_pnl)
            ep = len(self.episodes)
            if ep % self.log_every == 0:
                print(f"[Episode {ep:>4}] PnL: {self.ep_pnl:8.2f}")
            self.ep_pnl = 0              # reset for next ep
        return True

    def _on_training_end(self):
        pnl_series = pd.Series(self.episodes, name="episode_pnl")
        pnl_series.to_csv("episode_pnl_log.csv", index=False)
        print("Saved episode PnL to episode_pnl_log.csv")

def linear_schedule(initial_value: float):
        def func(progress_remaining: float) -> float:
            return progress_remaining * initial_value
        return func

model = PPO(
    "MlpPolicy", 
    train_env,
    seed=SEED,
    learning_rate=lr_schedule,
    n_steps=2048,           # collect more steps before update
    batch_size=64,         # larger batch for smoother gradients
    verbose=1,
    tensorboard_log="./ppo_tensorboard/",
    normalize_advantage=True,
    gae_lambda=0.6,
    gamma=0.95,
    ent_coef=0.05,
    clip_range=0.3,
    target_kl=0.05,
    max_grad_norm=0.8,

    vf_coef=0.25
)


# Train
#model.learn(total_timesteps=200000, tb_log_name="first_run")
metrics_cb = TradeLogger(log_every=20)
model.learn(total_timesteps=200000,
            tb_log_name="first_run", callback=metrics_cb)

print("Episodes logged:", len(metrics_cb.episodes))
           
# âœ… Save normalization stats after training
train_env.save("vecnormalize_stats.pkl")
print("Saved normalization stats to vecnormalize_stats.pkl")


test_env = VecNormalize.load("vecnormalize_stats.pkl", test_env_raw)
test_env.training = False
test_env.norm_reward = False

obs = test_env.reset()
ppo_actions = []
ep_pnl = 0
wins = 0
losses = 0
episodes = 0
equity_curve = []
step_counter = 0

trade_wins   = 0
trade_total  = 0 

while True:
    action, _ = model.predict(obs, deterministic=False)
    obs, reward, done, info = test_env.step(action)
    
    # Track action
    ppo_actions.append(int(action[0]))

    # ------ realised PnL tracking ----------------------------------
    realised = info[0]["trade_pnl"]          # 0.0 except when a trade closes
    ep_pnl  += realised

    # âœ¨ count trades & wins ----------------------------------------
    if realised != 0.0:                      # a trade just closed
        trade_total += 1
        if realised > 0:
            trade_wins += 1

    # Track PnL
    #ep_pnl += reward[0]
    #ep_pnl += info[0]["trade_pnl"]
    
    step_counter += 1

    if done[0]:                              # episode finished
        equity_curve.append(ep_pnl)
        ep_pnl  = 0.0
        episodes += 1
        obs = test_env.reset()

        # Stop if all steps in test set have been consumed
        if step_counter >= len(test_df):
            break

ppo_actions = np.array(ppo_actions)


# total_closed = wins + losses
# win_rate = wins / (total_closed + 1e-8)   # avoid div-by-zero

win_rate = trade_wins / trade_total if trade_total else 0.0
print(f"Per-trade win-rate     : {win_rate:.3f} ({trade_wins}/{trade_total})")
print(f"Mean episode PnL       : {np.mean(equity_curve):.2f}")

plt.plot(np.cumsum(equity_curve))
plt.title("Equity curve (test episodes)")
plt.show()



# ---- Baseline Comparison ----
from sklearn.metrics import classification_report

#Generate ground-truth labels (Buy=1, Sell=2, Hold=0)
THRESH = 0.008
LOOKAHEAD  = 2

#gt_labels = []
gt_labels = generate_ground_truth_actions(
                test_df_raw,
                threshold = THRESH,
                lookahead = LOOKAHEAD
            ).tolist()

# First PPO action has no ground truth, so remove it
gt_trimmed = gt_labels[:len(ppo_actions)]

uniq_labels = sorted(set(gt_trimmed) | set(ppo_actions[:len(gt_trimmed)]))
name_map    = {0: "Hold", 1: "Buy", 2: "Sell"}
target_names_dynamic = [name_map[l] for l in uniq_labels]

print(classification_report(
        gt_trimmed,
        ppo_actions[:len(gt_trimmed)],
        labels=uniq_labels,
        target_names=target_names_dynamic,
        zero_division=0))


# ---- Final Evaluation Metrics ----
def evaluate_final_metrics(equity_curve):
    equity_curve = np.array(equity_curve)
    returns = np.diff(equity_curve, prepend=0)

    peak = equity_curve[0]
    drawdowns = []

    for x in equity_curve:
        peak = max(peak, x)
        drawdowns.append(peak - x)

    max_drawdown = max(drawdowns) if drawdowns else 0


    sharpe = np.mean(returns) / (np.std(returns) + 1e-8) if len(returns) > 1 else 0.0

    print("\n--- Final Evaluation Metrics ---")
    print(f"Mean Return           : {np.mean(returns):.4f}")
    print(f"Std Dev of Returns    : {np.std(returns):.4f}")
    print(f"Sharpe Ratio Approx.  : {sharpe:.4f}")
    print(f"Max Drawdown          : {max_drawdown:.2f}")
    print(f"Win Rate              : {win_rate:.3f} "
          f"({trade_wins}/{trade_total + 1e-8})")
         # f"({wins}/{total_closed + 1e-8})")

evaluate_final_metrics(equity_curve)
```
